```python
import time
import math
```

```python
n = set(['1','3','7','9'])
def check1379(i):
    s = str(i)
    for j in range(len(s)):
        if s[j] not in n:
            return False
    return True

def isPrime(i):
    if(i == 1):
        return False
    for j in range(2,int(math.sqrt(i) + 2)):
        if(i % j == 0):
            return False
    return True

def truncateLeft(i):
    for j in range(len(str(i))):
        if(not(isPrime(i))):
            return False
        else:
            i = i % (10 ** (len(str(i)) - 1))
    return True
    
def truncateRight(i):
    for j in range(len(str(i))):
        print i
        if(not(isPrime(i))):
            return False
        else:
            i = i/10
    return True
```

```python
start = time.time()

l = []
i = 13
while(len(l) != 11):
    if(check1379(i)):
        if(truncateLeft(i) and truncateRight(i)):
            l.append(i)
            print i
        
    i += 4

print l, sum(l)
print "Runtime: %.3f seconds" % (time.time() - start)
```
