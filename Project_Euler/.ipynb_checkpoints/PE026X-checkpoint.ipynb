```python
import time
```

```python
def findCycleLength(f):
    i = 0
    s = "{0:.50f}".format(f)
    print s
    for j in range(20):
        f = f * 10
        print f
        f = f - int(f)
        print f
        for k in range(20):
            break
    return i

def twoAndFiveDivisible(i):
    if(i == 1):
        return True
    if(i % 2 == 0):
        return twoAndFiveDivisible(i/2)
    elif(i % 5 == 0):
        return twoAndFiveDivisible(i/5)
    else: return False
    
```

```python
start = time.time()

result = 0
longest_cycle = 0
for i in range(3,1000):
    if(not(twoAndFiveDivisible(i))):
        cycle = findCycleLength(1.0/i)
        if(cycle == 0):
            print "error for", i
        if(cycle > longest_cycle):
            longest_cycle = cycle
            result = i
print result
        

print "Runtime: %.3f seconds" % (time.time() - start)
```
